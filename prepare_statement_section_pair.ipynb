## Manually change some words

### Opus-MT

Opus NMT translates 'trial' as '审判', which leans to 'judgement' rather than 'experiment' (in train, dev, test.json files)

```python
def modify_statements_in_file(input_file_path, output_file_path):
    with open(input_file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)

    # 遍历所有的键（每个键对应一个记录）
    for record_id in data:
        if 'Statement' in data[record_id]:
            # 替换 'Statement' 字段中的 '审判' 为 '试验'
            data[record_id]['Statement'] = data[record_id]['Statement'].replace('审判', '试验')

    with open(output_file_path, 'w', encoding='utf-8') as file:
        json.dump(data, file, ensure_ascii=False, indent=4)

```

```python
def process_files(folder_path):
    target_files = ['train.json', 'dev.json', 'test.json']
    for entry in os.listdir(folder_path):
        input_path = os.path.join(folder_path, entry)
        if os.path.isfile(input_path) and entry in target_files:
            output_file_name = os.path.splitext(entry)[0] + '_modification.json'
            output_path = os.path.join(folder_path, output_file_name)
            modify_statements_in_file(input_path, output_path)
```

```python
# 使用 process_files 函数处理指定的文件夹
folder_path = './nli4ct_opus/'  
process_files(folder_path)
```

手动把文件移了位置，这样可以和修改的CT_json放一块

Opus NMT translates 'Interventions' as '间断', which is wrong meaning. I followed the English-Chinese contents in https://www.chictr.org.cn/showproj.html?proj=228156 and decided manually modified it as '干预措施' (in CT_json files)

```python
def replace_text_in_json(input_file_path, output_file_path):
    # 读取 JSON 数据
    with open(input_file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
    
    # 检查 "Intervention" 键是否在数据中
    if "Intervention" in data:
        # 替换 "间断" 为 "干预措施"
        data["Intervention"] = [item.replace("间断", "干预措施") for item in data["Intervention"]]
    
    # 确保输出目录存在
    os.makedirs(os.path.dirname(output_file_path), exist_ok=True)
    
    # 将修改后的数据写回到新文件
    with open(output_file_path, 'w', encoding='utf-8') as file:
        json.dump(data, file, ensure_ascii=False, indent=4)
```

```python
def process_CT_folder(input_folder_path, output_folder_path):
    # 遍历文件夹中的所有文件
    for filename in os.listdir(input_folder_path):
        if filename.endswith('.json'):
            input_file_path = os.path.join(input_folder_path, filename)
            output_file_path = os.path.join(output_folder_path, filename)
            replace_text_in_json(input_file_path, output_file_path)
            print(f'Processed and saved {filename} to {output_folder_path}')
```

```python
import json
import os

# List of keys representing the different dataset splits
key_json = ['train', 'dev', 'test']

# Ensure the directory for the output files exists
output_dir = './data/nli4ct_zh'
os.makedirs(output_dir, exist_ok=True)

# Process each key in key_json
for key in key_json:
    # List to hold the data for the current key
    statement_section_lst = []
    # Initialize an ID counter
    entry_id = 0
    
    # Open the corresponding JSON file for the current key
    with open(f"../nli4ct_opus_mod/{key}.json", 'r') as f:
        data_json = json.load(f)
        
        # Iterate through each entry in the JSON
        for k in data_json:
            line = data_json[k]
            
            # Load the primary premises
            with open(f"../nli4ct_opus_mod/CT_json/{line['Primary_id']}.json") as file:
                premises = json.load(file)
            # Initialize data structure for each entry with an ID
            current_data = {
                'id': entry_id,
                'hypothesis': line['Statement'],
                'CTR_premise': json.dumps(premises[line['Section_id']], ensure_ascii=False)
            }

            # Add the 'Label' only if it's not the test dataset
            if key != 'test':
                current_data['Label'] = line['Label']

            # Check for and process any secondary premises
            if 'Secondary_id' in line:
                with open(f"../nli4ct_opus_mod/CT_json/{line['Secondary_id']}.json") as sec_file:
                    secondary_premises = json.load(sec_file)
                combined_premises = json.loads(current_data['CTR_premise']) + secondary_premises[line['Section_id']]
                current_data['CTR_premise'] = json.dumps(combined_premises, ensure_ascii=False)

            # Append the processed data to the list
            statement_section_lst.append(current_data)
            # Increment the ID counter
            entry_id += 1

    # Save the processed data to a new JSON file
    with open(f"{output_dir}/{key}.json", 'w', encoding='utf-8') as outfile:
        json.dump(statement_section_lst, outfile, ensure_ascii=False, indent=4)
```

```python
def remove_quote(text):
    # Remove unnecessary escaped quotes
    text = text.replace('\"', '')  # Remove quotes
    return text
```

```python
def remove_space(text):
    # Regular expression to replace multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text)
    return text
```

##### remove unnecessary commas after colons and full points

```python
def remove_colons(text, language):
    text = text.replace(':,', ':')
    if language == 'en':
        text = text.replace('.,', '.')
    elif language == 'zh':
        text = text.replace('。,', '.')
    return text
```

```python
base_dir = './data/nli4ct_zh'

# Choose language: 'en' for English, 'zh' for Chinese
language = 'zh'

for key in key_json:
    with open(f"{base_dir}/{key}.json", 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    # Clean each entry
    for entry in data:
        entry['CTR_premise'] = remove_colons(entry['CTR_premise'], language='zh')

    # Overwrite the original JSON file with the cleaned data
    with open(f"{base_dir}/{key}.json", 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
```
